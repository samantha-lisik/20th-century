3. Install ChromeDriver

```python
# Selenium setup for macOS with Selenium 3.141.0

from selenium import webdriver
from selenium.webdriver.chrome.options import Options

chromedriver_path = "/Users/samlisik/Documents/Chromedriver/chromedriver"

# Set Chrome options
options = Options()
options.add_argument("--headless")     # run without opening a window
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")

driver = webdriver.Chrome(executable_path=chromedriver_path,
                          chrome_options=options)

# Test that it works
driver.get("https://www.google.com")
print("SUCCESS! Selenium + Chromedriver are working.")
```

Note on Selenium setup verification on macOS:

macOS blocked chromedriver initially ("cannot be opened because Apple cannot check it for malicious software").
I resolved this by:
1. Going to System Settings → Privacy & Security → "Allow Anyway"
2. Running the cell again and selecting "Open" when prompted

After this approval, Selenium successfully launched Chrome in headless mode.

The print message confirms the driver works.

3. Scraping the "Key events of the 20th century" wikipedia page.

```python
# Get the page's contents
url = "https://en.wikipedia.org/wiki/Key_events_of_the_20th_century"
driver.get(url)
```

```python
from selenium.webdriver.common.by import By

content_div = driver.find_element(By.ID, "mw-content-text")
page_text = content_div.text

print(page_text[:500])   # verify it worked
```

```python
# Save scraped content as a .txt file
filename = "key_events_20th_century.txt"

with open(filename, "w", encoding="utf-8") as f:
    f.write(page_text)

print(f"Saved scraped content to: {os.path.abspath(filename)}")
```
